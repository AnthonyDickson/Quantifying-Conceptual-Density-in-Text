```python
from collections import defaultdict
import math

import pandas as pd
```

```python
random_dfs = []

for i in range(1, 31):
    df = pd.read_csv('results/XMLParser-closures_annotations_random_%d.csv' % i, index_col=0)
    random_dfs.append(df)
```

```python
df_concat = pd.concat(random_dfs)
df_by_index = df_concat.groupby(df_concat.index)
mean = df_by_index.mean()
std = df_by_index.std()
```

```python
mean
```

```python
2 * std
```

```python
margin_of_error = 1.960 * std / math.sqrt(len(random_dfs))
margin_of_error
```

```python
graph_based_df = pd.read_csv('results/XMLParser-closures_annotations-graph_based.csv', index_col=0)
graph_based_df
```

```python
rule_based_df = pd.read_csv('results/XMLParser-closures_annotations-rule_based.csv', index_col=0)
rule_based_df
```

```python
both_df = pd.read_csv('results/XMLParser-closures_annotations-both.csv', index_col=0)
both_df
```

```python
merged_df = pd.concat([mean, graph_based_df, rule_based_df, both_df], 
          axis=1, 
          keys=['random', 'graph-based', 'rule-based', 'both'])
merged_df = merged_df.swaplevel(0, 1, axis=1)
merged_df = merged_df.sort_index(axis=1, level=0)
merged_df
```

```python
merged_df = merged_df[[
    'precision', 'recall', 'f1'
          ]].loc[[
    'A Priori Concepts', 'Emerging Concepts', 'Concepts Overall', 
    'Forward References', 'Backward References', 'References Overall', 
    'Overall Average'
]].reindex(['random', 'graph-based', 'rule-based', 'both'], axis=1, level=1)
merged_df
```

```python
std.loc['A Priori Concepts']
```

```python
def add_ci(column):
    if column.name[1] == 'random':
        return pd.Series(['%.3f ($\pm$ %.3f)' % (column.loc[row], std.loc[row][column.name[0]]) for row in column.index if 'References' not in row], index=column.index)            
    else:
        return column

merged_with_ci_df = merged_df.apply(add_ci, axis=0)
merged_with_ci_df
```

```python
latex_table = merged_with_ci_df.to_latex(    
    column_format='|l|r|r|r|r|r|r|r|r|r|r|r|r|',
    float_format=lambda x: '%.3f' % x,
    escape=False
)

lines = latex_table.splitlines()
lines.insert(8, '\midrule')
lines.insert(12, '\midrule')

print('\n'.join(lines))
```
