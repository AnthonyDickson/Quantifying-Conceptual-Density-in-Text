```python
from collections import defaultdict
import math

import pandas as pd
```

```python
parser = 'XMLParser'
```

```python
random_dfs = []

for i in range(1, 31):
    df = pd.read_csv('%s-closures_annotations-random_%d.csv' % (parser, i), index_col=0)
    random_dfs.append(df)
```

```python
df_concat = pd.concat(random_dfs)
df_by_index = df_concat.groupby(df_concat.index)
mean = df_by_index.mean()
std = df_by_index.std()
```

```python
mean
```

```python
2 * std
```

```python
margin_of_error = 1.960 * std / math.sqrt(len(random_dfs))
margin_of_error
```

```python
graph_based_df = pd.read_csv('%s-closures_annotations-graph_based.csv' % parser, index_col=0)
graph_based_df
```

```python
rule_based_df = pd.read_csv('%s-closures_annotations-rule_based.csv' % parser, index_col=0)
rule_based_df
```

```python
both_df = pd.read_csv('%s-closures_annotations-both.csv' % parser, index_col=0)
both_df
```

```python
merged_df = pd.concat([mean, graph_based_df, rule_based_df, both_df], 
          axis=1, 
          keys=['random', 'graph-based', 'rule-based', 'both'])
merged_df = merged_df.swaplevel(0, 1, axis=1)
merged_df = merged_df.sort_index(axis=1, level=0)
merged_df
```

```python
merged_df = merged_df[[
    'precision', 'recall', 'f1'
          ]].loc[[
    'A Priori Concepts', 'Emerging Concepts', 'Concepts Overall', 
    'Forward References', 'Backward References', 'References Overall', 
    'Overall Average'
]].reindex(['random', 'graph-based', 'rule-based', 'both'], axis=1, level=1)
merged_df
```

```python
def get_formatted_rows(column):
    rows = []
    
    for row in column.index:
        rows.append('%.3f ($\pm$ %.3f)' % (column.loc[row], margin_of_error.loc[row][column.name[0]]))
            
    return rows

def add_ci(column):
    if column.name[1] == 'random':
        return pd.Series(get_formatted_rows(column), index=column.index)            
    else:
        return column

merged_with_ci_df = merged_df.apply(add_ci, axis=0)
merged_with_ci_df
```

```python
# Order of first level columns gets messed up for some reason
merged_with_ci_df.columns = merged_with_ci_df.columns.set_levels([
    ['$\text{F}_1$ Score', 'Precision', 'Recall'], 
    ['Random', 'Graph-Based', 'Rule-Based', 'Both']
]) 
merged_with_ci_df
```

```python
for metric in ['Precision', 'Recall', '$\text{F}_1$ Score']:
    latex_table = merged_with_ci_df[metric].to_latex(    
        column_format='|l|r|r|r|r|',
        float_format=lambda x: '%.3f' % x,
        escape=False
    )

    lines = latex_table.splitlines()
    lines.insert(7, '\midrule')
    lines.insert(11, '\midrule')

    print('\n'.join(lines))
    print()
```
